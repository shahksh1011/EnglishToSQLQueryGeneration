```python

def pos_tokens_func(englishInputNormal,notstopwords):
    filtered_pos = []
    pos_tokens = tokenizer.tokenize(englishInputNormal)
    for p in pos_tokens:
        if p.lower() not in notstopwords:
            filtered_pos.append(p)
    return nltk.pos_tag(filtered_pos)
        
def get_proper_nouns_func(tagged_sent):
    return [word for word,pos in tagged_sent if pos == 'NNP']

def get_whPronouns_func(tagged_sent):
    return [word for word,pos in tagged_sent if pos == 'WP']

def get_commonNouns_func(tagged_sent):
    return [word for word,pos in tagged_sent if pos == 'NN']
```

```python
englishInput = input("What do you want?")
print(englishInput)

englishInputNormal = englishInput
englishInput=englishInput.lower()

tokenizer = RegexpTokenizer(r'\w+')
word_tokens=tokenizer.tokenize(englishInput)

# pos_tokens = tokenizer.tokenize(englishInputNormal)
# for p in pos_tokens:
#     if p.lower() not in notstopwords:
#         filtered_pos.append(p)

tagged_sent = pos_tokens_func(englishInputNormal, notstopwords)
print(tagged_sent)

propernouns = get_proper_nouns_func(tagged_sent)
whPronouns = get_whPronouns_func(tagged_sent)
commonNouns = get_commonNouns_func(tagged_sent)

print("****************")

for w in propernouns:
    print(w)
print("*****************")

stopWords = set(stopwords.words('english'))
wordsFiltered = []

filteredStopWords= []
for w in stopWords:
    if w not in notstopwords:
        filteredStopWords.append(w)
for w in word_tokens:
    print(w)
    w.replace("\\", "")
    if w not in filteredStopWords:
        wordsFiltered.append(w)
    
text2=nltk.Text(word_tokens)
bigrams = ngrams(text2,2)
print (list(bigrams))
lemmatizer = WordNetLemmatizer()
for w in wordsFiltered:
    print(lemmatizer.lemmatize(w))
```
